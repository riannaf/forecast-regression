## <h3> 6.2	Cek adanya missing value </h3>

```python
# the code below is adopted from this https://github.com/ashishpatel26/Amazing-Feature-Engineering/blob/master/feature_cleaning/missing_data.py

def check_missing(data,output_path=None):
    """
    check the total number & percentage of missing values
    per variable of a pandas Dataframe
    """

    result = pd.concat([data.isnull().sum(),data.isnull().mean()],axis=1)
    result = result.rename(index=str,columns={0:'total missing',1:'proportion'})
    if output_path is not None:
        result.to_csv(output_path+'missing.csv')
        print('result saved at', output_path, 'missing.csv')
    return result
```

# 8	Transformasi Data

## 8.1 Transformasi Data Kendaraan Bermotor
Karena kami tidak memiliki data bulanan maka kami mengunakan mean dari data tahunan.Namun memberikan nilai yang sama pada satu tahun itu sangat tidak realisitis sehingga kami perlu mentransformasi data menjadi lebih realistis dengan membuat data random yang memenuhi kriteria mean dari total kendaraan dan deviasi standar = 0.2 * mean.  

```python
# Tampilan menggunakan mean dari jumlah kendaraan bermotor
sns.regplot(x="Kendaraan_Bermotor", y="Kecelakaan", data=df)
```

**Gambar 21 Data Kendaraan Bermotor sebelum ditransformasi.**

```python
df[['Kendaraan_Bermotor','Kecelakaan']].corr()
```

```python
tahun = [12,12] # jumlah bulan pada setiap tahun
desired_mean_all = pd.unique(df['Kendaraan_Bermotor']) # temukan semua nilai mean untuk setiap tahun
data_baru = []
for num_bulan, mean in zip(tahun,desired_mean_all):
    num_samples = num_bulan
    desired_mean = mean
    desired_std_dev = desired_mean * 0.2
    np.random.seed(0)
    samples = np.random.normal(loc=0.0, scale=desired_std_dev, size=num_samples)
    actual_mean = np.mean(samples)
    actual_std = np.std(samples)
    zero_mean_samples = samples - (actual_mean)
    zero_mean_mean = np.mean(zero_mean_samples)
    zero_mean_std = np.std(zero_mean_samples)
    scaled_samples = zero_mean_samples * (desired_std_dev/zero_mean_std)
    scaled_mean = np.mean(scaled_samples)
    scaled_std = np.std(scaled_samples)
    final_samples = scaled_samples + desired_mean
    final_mean = np.mean(final_samples)
    final_std = np.std(final_samples)
    data_baru.extend(final_samples)

data_baru = np.array(data_baru).astype(np.int64)
df['Kend_baru'] = data_baru # gantikan data lama dengan data baru di kolom Kendaraan_Bermotor
```

```python
sns.regplot(x="Kend_baru", y="Kecelakaan", data=df)
```

**Gambar 22 Transformasi Data  Kendaraan Bermotor (Kendaraan_Bermotor) menjadi Kendaraan Bermotor baru (Kend_baru).**

```python
df[['Kend_baru','Kecelakaan']].corr()
```

Nilai korelasi menurun dari 0.252269 menjadi 0.110489

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle=False)
```

```python
regressors = [
    linear_model.LinearRegression(),
    RandomForestRegressor(n_estimators = 100),
    DecisionTreeRegressor()
    ]
pred_all = []
MAE_MSE_R2 = []

for regressor in regressors:
    pipe = Pipeline(steps=[
        ('classifier', regressor)])

    pipe.fit(X_train, y_train)
    predictions = pipe.predict(X_test)
    pred_all.append(predictions)
    error_r2 = [np.mean(np.absolute(y_test - predictions)), np.mean((y_test - predictions) ** 2), r2_score(y_test , predictions)]
    MAE_MSE_R2.append(error_r2)
    # print(classifier,"model score: %.3f" % pipe.score(X_test, y_test),'\n')
```
